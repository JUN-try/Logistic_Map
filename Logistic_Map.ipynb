```python
import numpy as np
import matplotlib.pyplot as plt
```

### ロジスティック方程式

```python
def q_func(a,x):
        return (a*x)*(1-x)
```

### ロジスティックマップ

```python

class Logistic_Map:
    def __init__(self,a=2.8,x0=0.123,switch="all",plot=500):
        
        #初期条件の設定
        self.a=a
        self.x0=x0
        self.switch=switch
        self.plot=plot
        
   
    def drawing(self,on_off="on"):
    
        X=[self.x0]
        Y=[self.x0]

        #ロジマップ
        for i in range(self.plot):
    
            if(i%2==0):
            #(x0,x1)
                y=q_func(self.a,X[i])
        
                X.append(X[i])
                Y.append(y)
    
            #(x1,x1)
                X.append(y)
                Y.append(y)
        
        
        size=int(self.plot/2)
        
        if(self.switch=="short"):
            plt.plot(X[:-size],Y[:-size])
           
        elif(self.switch=="long"):
            plt.plot(X[-size:],Y[-size:])
            
        elif(self.switch=="all"):
            plt.plot(X,Y)
        
        
        if(on_off=="on"):
            #y=xのプロット
            x1=[i/10 for i in range(100)]    
            y1=x1
            plt.plot(x1,y1)

            #y=x*xのプロット
            x2=[i/10 for i in range(100)]    
            y2=[ q_func(self.a,i) for i in x2]
            plt.plot(x2,y2)
    
        #x,yの範囲
        plt.xlim(0,1)
        plt.ylim(0,1)
        
        #ラベルとタイトル
        plt.title("Iteration Diagram of Logistic Map")
        plt.xlabel("Xn")
        plt.ylabel("Xn+1")
        
        #グラフの描画
        plt.show() 
```

### a=2.8 x0=0.123

```python
log1=Logistic_Map()
log1.drawing()
```

### a=3.83 x0=0.123

```python
log2=Logistic_Map(a=3.83,switch="long")
log2.drawing()
```

###  a=3.9 x0=0.123

```python
log3=Logistic_Map(a=3.9,switch="long")
log3.drawing(on_off=None)
```
